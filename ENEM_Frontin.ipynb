```python
print("Significado dos atributos/colunas:")
print("   - CO_PROVA_CN: Código do tipo de prova de Ciências da Natureza")
print("   - CO_PROVA_CH: Código do tipo de prova de Ciências Humanas")
print("   - CO_PROVA_LC: Código do tipo de prova de Linguagens e Códigos")
print("   - CO_PROVA_MT: Código do tipo de prova de Matemática")
print("   - NU_NOTA_CN: Nota da prova de Ciências da Natureza")
print("   - NU_NOTA_CH: Nota da prova de Ciências Humanas")
print("   - NU_NOTA_LC: Nota da prova de Linguagens e Códigos")
print("   - NU_NOTA_MT: Nota da prova de Matemática")
print("   - NU_NOTA_COMP1: Nota da Competência 1 da Redação")
print("   - NU_NOTA_COMP2: Nota da Competência 2 da Redação")
print("   - NU_NOTA_COMP3: Nota da Competência 3 da Redação")
print("   - NU_NOTA_COMP4: Nota da Competência 4 da Redação")
print("   - NU_NOTA_COMP5: Nota da Competência 5 da Redação")
print("   - NU_NOTA_REDACAO: Nota total da Redação")
```

```python
# Lista de colunas desejadas
colunas_desejadas = [
    'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT',
    'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT'
]

# Calculando estatísticas
estatisticas = df_filtrado[colunas_desejadas].describe().loc[['mean', 'std', '50%', 'min', 'max']]

# Renomeando linhas para português
estatisticas.index = ['Média', 'Desvio Padrão', 'Mediana', 'Mínimo', 'Máximo']

# Exibindo resultados
print(estatisticas)
```

```python
import matplotlib.pyplot as plt
# Lista das colunas de notas
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']

# Lista das colunas de códigos de prova
colunas_codigos = ['CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT']

# Histograma das notas
df_filtrado[colunas_notas].hist(bins=20, figsize=(12, 8))
plt.suptitle('Distribuição das Notas por Área', fontsize=16)
plt.tight_layout()
plt.show()

# Gráficos de barras com a contagem dos códigos de prova
for coluna in colunas_codigos:
    plt.figure()
    df_filtrado[coluna].value_counts().sort_index().plot(kind='bar')
    plt.title(f'Distribuição dos Códigos de {coluna}')
    plt.xlabel('Código da Prova')
    plt.ylabel('Número de Alunos')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
medias = df[colunas_notas].mean()
medias.plot(kind='bar', figsize=(8, 5), title='Média das Notas por Área')
plt.ylabel('Nota Média')
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df[colunas_notas].corr(), annot=True, cmap='Blues')
plt.title('Correlação entre Notas')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Colunas de nota
#colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']

# Boxplot
df_melt = df[colunas_notas].melt(var_name='Área', value_name='Nota')
plt.figure(figsize=(10,6))
sns.boxplot(x='Área', y='Nota', data=df_melt)
plt.title('Distribuição das Notas - Caxias')
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Colunas de nota
#colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']

# Boxplot
df_melt = df[colunas_codigos].melt(var_name='Área', value_name='Nota')
plt.figure(figsize=(10,6))
sns.boxplot(x='Área', y='Nota', data=df_melt)
plt.title('Distribuição - Caxias')
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Filtra alunos com notas válidas em CN, MT e Redação
df_plot = df[['NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].dropna()

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Eixos
x = df_plot['NU_NOTA_CN']
y = df_plot['NU_NOTA_MT']
z = df_plot['NU_NOTA_REDACAO']

# Dispersão
ax.scatter(x, y, z, c=z, cmap='viridis', s=20)

# Rótulos
ax.set_xlabel('Nota CN')
ax.set_ylabel('Nota MT')
ax.set_zlabel('Nota Redação')
ax.set_title('Dispersão 3D das Notas')

plt.show()
```

```python
colunas_redacao = [
    'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
    'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO'
]

# Estatísticas
estat_redacao = df_filtrado[colunas_redacao].describe().loc[['mean', 'std', '50%', 'min', 'max']]
estat_redacao.index = ['Média', 'Desvio Padrão', 'Mediana', 'Mínimo', 'Máximo']

print("\nEstatísticas das notas da redação e competências:")
print(estat_redacao)
```

```python
# Histogramas das competências e nota final da redação
df_filtrado[colunas_redacao].hist(bins=20, figsize=(12, 8))
plt.suptitle('Distribuição das Notas de Redação e Competências', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Se quiser ver a distribuição exata de notas por competência
for coluna in colunas_redacao:
    plt.figure()
    df_filtrado[coluna].value_counts().sort_index().plot(kind='bar')
    plt.title(f'Distribuição das Notas - {coluna}')
    plt.xlabel('Nota')
    plt.ylabel('Número de Alunos')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```
